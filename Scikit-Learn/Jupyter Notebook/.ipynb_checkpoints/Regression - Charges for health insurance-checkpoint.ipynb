# Data Transformation

```python
del df[ 'region' ] # A variavel smoker,... by AndrÃ© Batista
# A variavel smoker, iremos recodifica-la, de modo que yes serÃ¡ 1 e no serÃ¡ 0
df['smoker'] = df['smoker'].apply(lambda x : 1 if x == 'yes' else 0)

# faremos o mesmo para o sexo, com 1 para masc e 2 para feminino
df['sex'] = df['sex'].apply(lambda x : 1 if x == 'male' else 2)
```

```python
df.charges.max()
```

```python
df.charges.quantile(q=.99)
```

```python
df[df.charges >= df.charges.quantile(q=.68)]
```

```python
df = df[df.charges <= df.charges.quantile(q=.95)]
```

```python
sns.boxplot(df.charges)
```

# Modeling

- R2
- RMSE

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
# Import library for metrics
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn import linear_model

X = df.drop(columns='charges')
y = df.charges

X_train, X_test, y_train, y_test = train_test_split(X, y)
```


### Linear Regression

all linear regression models need to have the predictors scaled

```python
lin_reg = linear_model.LinearRegression()
lin_reg.fit(X_train, y_train)
y_lin_reg = lin_reg.predict(X_test)
```

```python
lin_reg.coef_
```

```python
lin_reg.feature_names_in_
```

```python
# Mean absolute error (MAE)
mae = mean_absolute_error(y_test, y_lin_reg)
# Mean squared error (MSE)
mse = mean_squared_error(y_test, y_lin_reg)
# R-squared scores
r2 = r2_score(y_test, y_lin_reg)

# Print metrics
print('Mean Absolute Error:', round(mae, 2))
print('Mean Squared Error:', round(mse, 2))
print('Root Mean Squared Error:', round(math.sqrt(mse), 2))
print('R-squared scores:', round(r2, 2))
```

### Random Forest

```python
rf = RandomForestRegressor(max_depth=3, random_state=0, n_estimators=50)
rf.fit(X_train, y_train)
```

```python
y_pred_rf = rf.predict(X_test)
```

```python
mean_squared_error(y_test, y_pred_rf)
```

```python
rf.feature_importances_
```

```python
rf.feature_names_in_
```

```python
# Mean absolute error (MAE)
mae = mean_absolute_error(y_test, y_pred_rf)
# Mean squared error (MSE)
mse = mean_squared_error(y_test, y_pred_rf)
# R-squared scores
r2 = r2_score(y_test, y_pred_rf)

# Print metrics
print('Mean Absolute Error:', round(mae, 2))
print('Mean Squared Error:', round(mse, 2))
print('Root Mean Squared Error:', round(math.sqrt(mse), 2))
print('R-squared scores:', round(r2, 2))
```
