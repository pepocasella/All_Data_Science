# Gaussian Distribution Phenomenon

Examples of observations that fit a Gaussian distribution

- Body temperature
- People’s Heights
- Car mileage
- IQ scores

Some examples of observations that do not fit a Gaussian distribution and instead may fit an exponential (hockey-stick shape) include:

- People’s incomes
- Population of countries
- Sales of cars.


### The ideal bell-shaped curve of the distribution

```python
## generate the data and plot it for an ideal normal curve
## x-axis for the plot
x_data = np.arange(-5, 5, 0.0001)

## y-axis as the gaussian
y_data = stats.norm.pdf(x_data, 0, 1) # Probability density function evaluated at x

## plot data
plt.plot(x_data, y_data)
plt.title('Bell curve which is also called the Gaussian distribution a.k.a. normal distribution.') # also know as
plt.show()
```

```python
mu = 0 
sigma = 5
d1 = np.random.normal(mu - 200, sigma*5, [100000,1]).ravel() 
d2 = np.random.normal(mu + 20, sigma*15, [100000,1]).ravel() 
d3 = np.random.normal(mu+ 160, sigma*25, [100000,1]).ravel() 

## plot data
plt.hist(d1, alpha=0.5, bins=100, color='red', label='{mu - 200, sigma*5}')
plt.hist(d2, alpha=0.5, bins=100, color='blue', label='{mu + 20, sigma*15}')
plt.hist(d3, alpha=0.5, bins=100, color='green', label='{mu + 160, sigma*25}')
plt.title('Gaussian Distribution') # also know as
plt.legend()
plt.show()
```

```python
##setting the seed for the random generation
np.random.seed(1)
##generating univariate data
data = 10 * np.random.randn(1000000) + 100 # sigma * np.random.randn(...) + mu

fig, axes = plt.subplots(2,1,figsize=(16,8))
axes[0].hist(data, bins=100)
axes[1].boxplot(data, vert=False)
plt.show()
```

```python
mu = 100
sigma = 10
noise = np.random.normal(mu, sigma, [100000,1]).ravel() 

fig, axes = plt.subplots(2,1,figsize=(16,8))
axes[0].hist(noise, bins=100)
axes[1].boxplot(noise, vert=False)
plt.show()
```
